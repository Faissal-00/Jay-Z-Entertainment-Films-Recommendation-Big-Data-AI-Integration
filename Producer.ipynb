```python
import requests
import json

api_url = 'http://127.0.0.1:5000/movies/page{}'  # Replace with your API endpoint

def get_movie_data(page_number):
    response = requests.get(api_url.format(page_number))
    
    if response.status_code == 200:
        movie_data = response.json()
        return movie_data
    else:
        print(f"Failed to fetch data from page {page_number}")
        return None

# Example: Get data from page 1
page_number = 1
data_from_api = get_movie_data(page_number)

if data_from_api:
    print(f"Received data from page {page_number}:")
    print(json.dumps(data_from_api, indent=2))
```
